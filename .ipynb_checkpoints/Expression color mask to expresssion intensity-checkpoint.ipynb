```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.cm as cm
import numpy as np
import scipy.cluster.vq as scv

def colormap2arr(arr,cmap, cutoff=1.0):    
    # Modified from http://stackoverflow.com/questions/3720840/how-to-reverse-color-map-image-to-scalar-values/3722674#3722674
    N = 64
    gradient = np.zeros((N,4))
    gradient[1:,:]=cmap(np.linspace(0.0,1.0,N-1))
    if cmap==cm.jet: ## for jet color map, the last few entries are closer to origin
        maxRed=np.argwhere(gradient[:,0][::-1]>=cutoff)[0][0] 
        gradient = gradient[:-maxRed,:] ## remove those before mapping    
    # Reshape arr to N by 4
    arr2=arr.reshape((arr.shape[0]*arr.shape[1],arr.shape[2]))
    # Use vector quantization to shift the values in arr2 to the nearest point in
    # the code book (gradient).
    code,dist=scv.vq(arr2,gradient)
    # Scale the values so they are from 0 to 1.
    values=code.astype('float')/gradient.shape[0]
    # Reshape values back to original shape
    values=values.reshape(arr.shape[0],arr.shape[1])
    return values

def prepArr(arr):
    arr4c = np.ones((arr.shape[0],arr.shape[1],4))
    arr_copy = arr.copy()/255.0 ## normalize to range(0, 1.0)
    arr4c[:,:,:3] = arr_copy ## return 4 channel array
    return arr4c

def rgb2intensity(arr,cmap):
    arr4c = prepArr(arr)    
    return colormap2arr(arr4c,cmap)
```

## Load data
data were saved from original ISH images and expression mask downloaded using Allen Brain institute's APIs (https://help.brain-map.org/display/api/Allen+Brain+Atlas+API)

```python
import pickle
imgeID = 'Gad1_sec148_75457539'
with open(imgeID+'_expression.pickle','rb') as handle:
    expressionVal = pickle.load(handle)

with open(imgeID+'_fish.pickle','rb') as handle:
    fishVal = pickle.load(handle)
```

## Have a quick look at the ISH data

```python
expression2D = np.prod(expressionVal, axis=2);
expressionPixels = (expression2D>0) ## choose among "expressing" pixels
```

```python
N = 100000
idx = np.random.choice(np.sum(expressionPixels), N); ## subsample N pixels 
xdataF = fishVal[expressionPixels,0][idx]
ydataF = fishVal[expressionPixels,1][idx]
zdataF = fishVal[expressionPixels,2][idx]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
Rprime = 256-xdataF
Gprime = 256-ydataF
Bprime = 256-zdataF
ax.scatter3D(Rprime, Gprime,Bprime, c='k',s=1.0);
ax.set_xlabel('255-R');
ax.set_ylabel('255-G');
ax.set_zlabel('255-B');
# ax.view_init(30, 145)
```

ISH image data points are almost in the diagonal plane.

```python
from sklearn import linear_model
```

```python
fig = plt.figure();
RG = np.sqrt(2)*((Rprime+Gprime)/2); # if we project R, G axes into diagonal plane
B = Bprime;

ransac = linear_model.RANSACRegressor(); 
ransac.fit(RG.reshape(-1,1),B.reshape(-1,1));
slope = ransac.estimator_.coef_[0][0];
intercept = ransac.estimator_.intercept_[0]
x2 = np.linspace(1, 350, 100);
y2 = ransac.predict(x2.reshape(-1,1));

plt.figure();
plt.plot(RG, B,'r.',alpha=0.3);
plt.plot(x2,y2,'b');
plt.xlabel('RG');
plt.ylabel('B');
plt.title(f'slope:{slope:.2f}, Intercept:{intercept:.2f}, N={N}');
```

Within the plane, it is almost linear with a slope of 0.68(This may not be true for other images). 

## We may as well predict Blue values from R, G directly

```python
RG2d = np.column_stack((Rprime,Gprime))

ransac = linear_model.RANSACRegressor(); 
ransac.fit(RG2d,B.reshape(-1,1));

bpredict = ransac.predict(RG2d);
```

```python
plt.figure();
ax = plt.axes(projection='3d')
ax.scatter3D(Rprime, Gprime,bpredict, c='k',s=1.0);
ax.scatter3D(Rprime, Gprime,Bprime, c='r',s=1.0,alpha=0.1);
plt.xlabel('RG');
plt.ylabel('B');
plt.title(f'coefficients:{ransac.estimator_.coef_[0][0]:.2f}, {ransac.estimator_.coef_[0][1]:.2f}, Intercept:{intercept:.2f}, N={N}');
```

The coefficents for the R and G axies came very close to the magic conversion equation mentioned in `Remarks` section at the end:

$0.21 R + 0.72 G + 0.07 B$

## Convert from RGB space back to intensity

```python
imgVals = rgb2intensity(expressionVal, cm.jet)
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,6))
axes = axes.ravel()
axes[0].imshow(fishVal,origin='upper')
axes[1].imshow(expressionVal,origin='upper')
axes[2].imshow(imgVals,origin='upper', cmap='jet')
for a in axes:
    a.axis('off')
plt.show()
```

## Now we have the intensity, we could finally find the full linear conversion coefficients:

```python
from skimage.filters import gaussian
from skimage.filters import median
from skimage.morphology import disk, area_closing
from skimage.filters.rank import maximum
```

```python
RGB2d = (255-fishVal)[expressionPixels,:]

targetVal = imgVals[expressionPixels]
```

```python
plt.hist(targetVal)
```

```python
np.min(targetVal), np.max(targetVal)
```

```python
ransac = linear_model.RANSACRegressor(); 
ransac.fit(RGB2d,targetVal.reshape(-1,1));

intensityPredict = ransac.predict(RGB2d);
ransac.estimator_.coef_
```

```python
intensityPredict2 = intensityPredict-np.min(intensityPredict)
```

```python
plt.hist(intensityPredict2)
```

```python
np.min(intensityPredict2),np.median(intensityPredict2),np.mean(intensityPredict2),np.max(intensityPredict2)
```

```python
predictImg = np.zeros_like(imgVals)
intensityPredict2 += np.median(intensityPredict) ## enhance contrast relative to background (non-expressing pixels)
for idx, val in zip(np.argwhere(expressionPixels),intensityPredict2):
    predictImg[idx[0],idx[1]]= val
```

```python
predictImg = median(predictImg, disk(3)) 
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,6))
axes = axes.ravel()
axes[0].imshow(fishVal,origin='upper')
axes[1].imshow(expressionVal,origin='upper')
axes[2].imshow(imgVals,origin='upper', cmap='jet')
axes[3].imshow(predictImg,origin='upper', cmap='jet')
for a in axes:
    a.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,12))
axes = axes.ravel()
axes[0].imshow(fishVal,origin='upper')
axes[1].imshow(expressionVal,origin='upper') ## original mask
axes[2].imshow(imgVals,origin='upper', cmap='jet') #intensity  mapped back from original mask
axes[3].imshow(predictImg,origin='upper', cmap='jet') #intensity predicted from ISH images
for a in axes:
    a.set_xlim([4200, 4900])
    a.set_ylim([4900, 5600])
    a.axis('off')
plt.show()
```

Note that the original mask were calculated after segmentation. That's why it is most uniform within each segmented region.
